```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.image as mpimg
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.image as mpimg
from tensorflow.keras.optimizers import Adam
```

```python
# List of image paths and titles
extra_images = ['amer_sign2.png', 'amer_sign3.png', 'american_sign_language.PNG']
extra_titles = ['Sign 2', 'Sign 3', 'American Sign Language']

# Function to display each image separately
def display_extra_images(image_paths, titles):
    for img_path, title in zip(image_paths, titles):
        img = mpimg.imread('/content/drive/MyDrive/Assignment3_UMLT_2024/' + img_path)
        plt.figure(figsize=(5, 5))
        plt.imshow(img)
        plt.title(title)
        plt.axis('off')
        plt.show()

# Display the images
display_extra_images(extra_images, extra_titles)
```

```python
# Define the path to the dataset files in Google Drive
train_data_path = '/content/drive/MyDrive/Assignment3_UMLT_2024/sign_mnist_train.csv'
test_data_path = '/content/drive/MyDrive/Assignment3_UMLT_2024/sign_mnist_test.csv'

# Load the dataset
train_data = pd.read_csv(train_data_path)
test_data = pd.read_csv(test_data_path)

# Display the first few rows of the training data
print(train_data.head())

# Extract features and labels from the training data
X_train = train_data.iloc[:, 1:].values
y_train = train_data.iloc[:, 0].values

X_test = test_data.iloc[:, 1:].values
y_test = test_data.iloc[:, 0].values

# Reshape the data to fit the model input (32x32 images)
X_train = X_train.reshape(X_train.shape[0], 32, 32, 1)
X_test = X_test.reshape(X_test.shape[0], 32, 32, 1)

# Normalize the data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Display sample images of each letter
def display_samples(X, y):
    plt.figure(figsize=(15, 10))
    for i in range(26):
        plt.subplot(4, 7, i + 1)
        idx = np.where(y == i)[0]
        if len(idx) > 0:
            plt.imshow(X[idx[0]].reshape(32, 32), cmap='gray')
            plt.title(chr(i + 65))
        plt.axis('off')
    plt.show()

display_samples(X_train, y_train)

```

```python
# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Verify the split
print(f"Training set size: {X_train.shape[0]}")
print(f"Validation set size: {X_val.shape[0]}")
```

```python
print(X_train.shape)
print(X_val.shape)
```

```python
# Define the corrected CNN model
def create_corrected_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 1)),  # Updated input shape
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),  # Flatten the output from convolutional layers
        Dense(128, activation='relu'),
        Dense(25, activation='softmax')  # Assuming 25 classes for A-Z (excluding J and Z which require motion)
    ])
    return model

# Compile the model
corrected_model = create_corrected_model()
corrected_model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the corrected model
history_corrected = corrected_model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=32)

# Evaluate the corrected model
corrected_loss, corrected_accuracy = corrected_model.evaluate(X_test, y_test)
print(f"Corrected Model Accuracy: {corrected_accuracy}")
```

```python
# Check the total number of parameters in the model
corrected_model.summary()
```

```python
# Define the reduced CNN model
def create_reduced_model():
    model = Sequential([
        Conv2D(16, (3, 3), activation='relu', input_shape=(32, 32, 1)),  # Reduced filters
        MaxPooling2D((2, 2)),
        Conv2D(32, (3, 3), activation='relu'),  # Reduced filters
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(64, activation='relu'),  # Reduced units
        Dense(25, activation='softmax')  # Output layer
    ])
    return model

# Compile the reduced model
reduced_model = create_reduced_model()
reduced_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Check the total number of parameters
reduced_model.summary()
```

```python
# Compile the reduced model
reduced_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the reduced model
history_reduced = reduced_model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=32)
```

```python
# Evaluate the reduced model
reduced_loss, reduced_accuracy = reduced_model.evaluate(X_test, y_test)
print(f"Reduced Model Accuracy: {reduced_accuracy}")
```

```python
# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Plot training & validation accuracy values
axs[0].plot(history_reduced.history['accuracy'], label='Training Accuracy')
axs[0].plot(history_reduced.history['val_accuracy'], label='Validation Accuracy')
axs[0].set_title('Model Accuracy Over Epochs')
axs[0].set_xlabel('Epochs')
axs[0].set_ylabel('Accuracy')
axs[0].legend(loc='lower right')
axs[0].grid(True)

# Plot training & validation loss values
axs[1].plot(history_reduced.history['loss'], label='Training Loss')
axs[1].plot(history_reduced.history['val_loss'], label='Validation Loss')
axs[1].set_title('Model Loss Over Epochs')
axs[1].set_xlabel('Epochs')
axs[1].set_ylabel('Loss')
axs[1].legend(loc='upper right')
axs[1].grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Predict on the test set
y_pred = np.argmax(reduced_model.predict(X_test), axis=1)
```

```python
# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Check the shape of the confusion matrix
print("Confusion matrix shape:", cm.shape)

# Print unique classes
print("Unique classes in y_test:", np.unique(y_test))
print("Unique classes in y_pred:", np.unique(y_pred))

# Find the missing class by comparing with the full range of classes
all_classes = set(range(25))
missing_classes = all_classes - set(np.unique(y_test)) - set(np.unique(y_pred))
print("Missing class(es):", missing_classes)
```

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Define class labels, excluding the missing class (9) if it still applies
class_labels = [chr(i + 65) for i in range(25) if i != 9]

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_labels)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

```

```python
from sklearn.metrics import classification_report

# Generate and print the classification report
report = classification_report(y_test, y_pred, target_names=class_labels)
print("Classification Report:\n", report)
```

```python
# Calculate the accuracy for each class
class_accuracies = np.diag(cm) / np.sum(cm, axis=1)

# Compute the median accuracy
median_accuracy = np.median(class_accuracies)
print(f"Unbiased Median Accuracy: {median_accuracy}")
```

```python
mean_accuracy = np.mean(class_accuracies)
print(f"Overall Mean Accuracy: {mean_accuracy}")
```

```python
max_accuracy_index = np.argmax(class_accuracies)
highest_accuracy_letter = class_labels[max_accuracy_index]
print(f"Letter with Highest Accuracy: {highest_accuracy_letter} ({class_accuracies[max_accuracy_index]})")
```

```python
min_accuracy_index = np.argmin(class_accuracies)
lowest_accuracy_letter = class_labels[min_accuracy_index]
print(f"Letter with Lowest Accuracy: {lowest_accuracy_letter} ({class_accuracies[min_accuracy_index]})")
```

```python
cm_copy = cm.copy()
np.fill_diagonal(cm_copy, 0)
common_errors = cm_copy.argmax(axis=1)
error_counts = cm_copy.max(axis=1)

# Find the top 3 errors
top_3_errors_indices = np.argsort(error_counts)[-3:]
for idx in top_3_errors_indices:
    print(f"Most common misclassification: {class_labels[idx]} as {class_labels[common_errors[idx]]}, count: {error_counts[idx]}")
```

**Model Development and Selection Process**

**Model Architecture**

For this task, I developed a Convolutional Neural Network (CNN) model that was designed to be both efficient and effective for classifying American Sign Language (ASL) letters. The architecture of the final selected model is as follows:

**Input Layer:**

**Input shape:** (32, 32, 1) for grayscale images of 32x32 pixels.

**Convolutional Layers:**

**Conv2D Layer 1**: 16 filters, 3x3 kernel size, ReLU activation.
**MaxPooling2D Layer 1**: Pool size of 2x2.
**Conv2D Layer 2**: 32 filters, 3x3 kernel size, ReLU activation.
**MaxPooling2D Layer 2**: Pool size of 2x2.

**Flatten Layer**:

Converts the 2D matrix data into a 1D vector.
Fully Connected (Dense) Layers:

Dense Layer 1: 64 units, ReLU activation.
Output Layer: 25 units (one for each letter class), softmax activation for multiclass classification.

**Number of Parameters**

The selected model was designed with efficiency in mind to meet the constraint of having fewer than 500,000 total parameters. After careful adjustments, the final model has 80,217 parameters, which is well within the required limit. This allows the model to be lightweight and suitable for deployment in resource-constrained environments, such as mobile applications.

**Training Process**

The model was trained using the following settings:

**Optimizer**: Adam  is used for the adaptive learning rate capabilities.
**Loss Function**: Sparse categorical cross-entropy, appropriate for multi-class classification problems.
**Metrics**: Accuracy, which tracks the proportion of correct predictions during training.
**Batch Size**: 32, which balances the computational efficiency and learning stability.
**Epochs**: 10, chosen after observing the convergence of both training and validation accuracy.

Throughout the training process, both training and validation accuracy were monitored to ensure that the model was learning effectively and not overfitting. Early stopping criteria were not applied in this instance as the model showed continuous improvement across all epochs.

**Final Model Selection Rationale**

The final model was selected based on the following criteria:

**Parameter Constraints**: The model meets the client's requirement of having fewer than 500,000 parameters.

**Performance**: The model achieved high accuracy on both the validation and test sets, with a final test accuracy of 88.7%.

**Generalization**: The model shows good generalization to the unseen test set, indicating that it is not overfitting.

**Efficiency**: The reduced model, with its lightweight architecture, is ideal for deployment in environments with limited computational resources, such as mobile devices.
